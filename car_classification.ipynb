```python
train_set = df.drop(['class'],axis=1)
```

```python
label = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_set, label, test_size=0.2, random_state=42, stratify=label)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
import category_encoders as ce
#rom sklearn.preprocessing import OrdinalEncoder , OneHotEncoder
```

```python
encoder= ce.OrdinalEncoder(cols=['buying', 'maint','lug_boot', 'safety'])
encoder_2 = ce.OneHotEncoder(cols=['doors','persons'])
X_train = encoder.fit_transform(X_train)
X_train = encoder_2.fit_transform(X_train)
X_test = encoder.transform(X_test)
X_test = encoder_2.transform(X_test)
```

```python
gridSearchCV=GridSearchCV(estimator=DecisionTreeClassifier(),param_grid={'max_depth':range(1,10)},cv=5)
gridSearchCV.fit(X_train,y_train)
print("best prameters: ", gridSearchCV.best_params_)
print("best score" , gridSearchCV.best_score_)
best_model=gridSearchCV.best_estimator_
```

```python
best_model.fit(X_train,y_train)
best_model_pred=best_model.predict(X_test)
print("best model accuracy score", accuracy_score(y_test,best_model_pred))
```

```python
print("training score", best_model.score(X_train,y_train))
print("testing score",  best_model.score(X_test,y_test))
```

```python
print("""confusion matrix:

"""
,classification_report(y_test,best_model_pred))
```
